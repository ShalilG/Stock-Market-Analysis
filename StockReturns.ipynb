```python
import pandas as pd 
import yfinance as yf
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
#Load the data 
tick = 'AAPL', 'AMZN', 'META', 'NFLX', 'DIS', 'T', 'CVX', 'XOM', 'MSFT', 'GOOG', 'SBUX', 'BKNG', 'V', 'MA', 'AXP', 'KO', 'PEP', 'IBM', 'ADBE', 'TSLA', 'F', 'GM', 'MMM', 'JNJ', 'PFE', 'AMGN', 'LLY', 'TJX', 'TGT', 'WMT', 'COST', 'MAR', 'QCOM', 'AVGO', 'INTC', 'NVDA', 'BA', 'JPM', 'BAC', 'WFC', 'C', 'FIS', 'FTNT', 'ABBV', 'ABT', 'UNH', 'ALGN', 'INTU', 'AMD', 'MU', 'CVS', 'MRK', 'GILD', 'BMY', 'HD', 'LOW', 'GS', 'PG', 'GSK', 'BP'

#Show the data
df = yf.download(tick, start = '2018-8-3', end = '2022-8-3')
df = df[['Adj Close']]

df
```

```python
cols = df.columns 
```

```python
#Get the DSR
daily_simple_returns = df[[cols[0], cols[1], cols[2], cols[3], cols[4], cols[5], cols[6], cols[7], cols[8], cols[9], cols[10], cols[11], cols[12], cols[13], cols[14], cols[15], cols[16], cols[17], cols[18], cols[19], cols[20], cols[21], cols[22], cols[23], cols[24], cols[25], cols[26], cols[27], cols[28], cols[29], cols[30], cols[31], cols[32], cols[33], cols[34], cols[35], cols[36], cols[37], cols[38], cols[39], cols[40], cols[41], cols[42], cols[43], cols[44], cols[45], cols[46], cols[47], cols[48], cols[49], cols[50], cols[51], cols[52], cols[53], cols[54], cols[55], cols[56], cols[57], cols[58], cols[59]]].pct_change(1)
print(daily_simple_returns)
```

```python
#Show the correlation 
plt.subplots(figsize=(30,30))
sns.heatmap(daily_simple_returns.corr(), annot= True, fmt= '.2%')
```

```python
#Show the co-variance matrix
daily_simple_returns.cov()
```

```python
#Get the variance 
daily_simple_returns.var()
```

```python
#Get volatility 
daily_simple_returns.std() * 100
```

```python
#Show the average daily simple returns 
daily_simple_returns.mean() * 100
```

```python
annualized_returns = daily_simple_returns.mean() * 100 * 252
annualized_returns
```
