```python
import pandas as pd 
import numpy as np 
import yfinance as yf
import jinja2
from pycaret.regression import *
from sklearn.model_selection import train_test_split
```

```python
#Load the data 
tick = 'SPY'

#Show the data for Adj Close
asset = yf.download(tick, start = '1993-01-29', end = '2022-10-4')
asset.index = asset.index.tz_localize(None)
asset['Date'] = asset.index
asset
```

```python
df3 = asset.copy()

#Store the data in a feature data set call 'X' and add in the target column

X = np.array(df3[['Date', 'Open', 'Close']])

#Store the close price in an array in a variable called 'y'

y = df3['Close'].values

#Split the data into 85% training and 15% testing 

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=0, shuffle=False)
```

```python
#Get the train data and transform it to a dataframe
train_data = pd.DataFrame(x_train, columns=['Date', 'Open', 'Close'])
#Show the first 7 rows of data 
train_data.head(7)
```

```python
#Get the test data and transform it to a dataframe
test_data = pd.DataFrame(x_test, columns=['Date', 'Open', 'Close'])
#Show the first 7 rows of data 
test_data.head(7)
```

```python
#Initialize the setup 

s = setup(data=train_data, target='Close', session_id = 123, use_gpu=True)
```

```python
#Train on multiple models and compare them 
compare_models(sort='MSE')
```

```python
#Create the model 

en = create_model('en')
```

```python
#Get the predictions 
predict_model(en, data=test_data.drop(['Close'], axis=1))
```
