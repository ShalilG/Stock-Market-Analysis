```python
#Predict the future price of a stock using machine learning, Facebook Prophet and python 
```

```python
#Import the libraries
```

```python
import pandas as pd
import numpy as np 
import yfinance as yf
```

```python
df = yf.download('AAPL', start = '2012-07-17', end = '2022-11-1')
df
```

```python
df = df.reset_index()
df
```

```python
df = df[['Date', 'Close']]
df = df.rename(columns={'Date':'ds', 'Close':'y'})
df
```

```python
#Get the last 20 rows of data and store them into a new variable 

last = df[len(df)-20:]
last
```

```python
#Get all of the rows of data except the last 20 

df = df[:-20 ]
df
```

```python
from fbprophet import Prophet
```

```python
#Creating the Prophet object (model)
fbp = Prophet(daily_seasonality = True)

#Fit or train the model 

fbp.fit(df)
future = fbp.make_future_dataframe(periods=365)
forecast = fbp.predict(future)
```

```python
#Import the facebook prophet plotting libraries 
from fbprophet.plot import plot_plotly, plot_components_plotly

plot_plotly(fbp, forecast)
```

```python
#Show the models prediction for 2022-10-15
forecast[forecast.ds == '2022-10-27']['yhat']
```

```python
#Show the actual value of the stock 2022-10-15
last[last.ds == '2022-10-27']['y']
```
