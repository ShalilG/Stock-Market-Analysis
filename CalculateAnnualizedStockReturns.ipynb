```python
import pandas as pandas
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
#Load the data 
tick = 'META', 'AMZN', 'TSLA', 'GOOG', 'NFLX', 'AAPL', 'CRM', 'NVDA', 'SPY', 'VOO', 'WMT', 'KO', 'NKE', 'MSFT', 'VGT'

#Show the data
df = yf.download(tick, start = '2017-01-01', end = '2022-12-31')
df = df[['Adj Close']]

df
```

```python
#Get the simply daily returns of the assets

daily_simple_returns = df.pct_change()
daily_simple_returns
```

```python
#Calculate the annualized returns 

annualized_returns = daily_simple_returns.mean() * 252

#Show the annualized returns 

annualized_returns
```

```python
#Sort the data 

sorted_annualized_returns = annualized_returns.sort_values(ascending=False)
sorted_annualized_returns
```
